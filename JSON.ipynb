```python
import json

name =  {
  "library_name": "Central Tech Archive",
  "location": {
    "city": "San Francisco",
    "floor_plan": [1, 2, 3]
  },
  "featured_books": [
    {
      "id": "B001",
      "title": "Learning JSON",
      "author": {
        "first_name": "Alice",
        "last_name": "Smith"
      },
      "available": True,
      "tags": ["coding", "data", "web"]
    },
    {
      "id": "B002",
      "title": "Mastering Python",
      "author": {
        "first_name": "Bob",
        "last_name": "Jones"
      },
      "available": False,
      "tags": ["programming", "backend"]
    }
  ],
  "total_books": 1250
}
with open('name.json', 'w') as f:
    json.dump(name, f,indent=2)
```
