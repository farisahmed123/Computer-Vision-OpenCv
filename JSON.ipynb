```python
import json
camera = {
  "timestamp_received": "2026-02-01T15:45:00Z",
  "camera_id": "CAM-04-PESHAWAR",
  "sensors": {
    "motion_detection": {
      "active": True,
      "confidence_score": 0.92,
      "area_affected": "entrance_gate"
    },
    "environment": {
      "lux_level": 15,
      "visibility_percentage": 45,
      "lighting_status": "low_light"
    },
    "stream_health": {
      "latency_ms": 1200,
      "frame_rate": 14,
      "dropped_packets": 5
    }
  },
  "previous_state": "idle"
}
with open('camera.json', 'w') as f:
    json.dump(camera, f,indent=2)
```

```python
import json

# Practice loading the file
with open('camera.json', 'r') as f:
    camera = json.load(f)

def main():
    # 1. Handle basic info
    cam_id = camera.get('camera_id', 'Unknown')
    print(f"Analyzing Camera: {cam_id}")

    # 2. Extract nested data into variables (makes code cleaner!)
    sensors = camera['sensors']
    latency = sensors["stream_health"]["latency_ms"]
    motion_active = sensors['motion_detection']['active']
    visibility = sensors['environment']['visibility_percentage']
    lux = sensors['environment']['lux_level']

    # 3. Decision Tree Logic
    # CRITICAL: Check latency first. If it's too high, don't trust the sensors.
    if latency > 1000:
        print("CRITICAL STATE: SYSTEM_DELAY_WARNING")
        return # Stop the function here

    # 4. Handle Environmental Edge Cases
    if lux < 20:
        print("ACTION: ACTIVATE_INFRARED")

    # 5. Handle Contradictions
    if motion_active and visibility < 50:
        print("STATE: CAUTION - Potential Motion with Low Visibility")
    elif motion_active:
        print("STATE: ALERT - Motion Confirmed")
    else:
        print("STATE: System Clear / Idle")

main()
```
