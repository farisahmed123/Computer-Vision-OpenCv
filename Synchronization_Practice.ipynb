```python
import json
def load(filename):
  with open(filename,'r') as f:
    return json.load(f)

def find(score):
  data = load('simple_data.json')
  count = 0
  for i in data:
    if i['score'] >= score:
      count+=1
  return count

print(find(0.1))
```

```python
import json
user ={"101": "Abbas", "102": "Hussain"}
server = {"restricted_doors": ["Main", "Server Room"]}
acces = [
  {"user_id": "101", "time": "09:00", "door": "Main"},
  {"user_id": "102", "time": "09:15", "door": "Gallery"}
]
with open ('users.json','w') as f:
  json.dump(user,f)
with open ('security_settings.json','w') as f:
  json.dump(server,f)
with open ('access_logs.json','w') as f:
  json.dump(acces,f)
```

```python
import json

def load_files(*filenames):
    return [json.load(open(f, 'r')) for f in filenames]

def analyze_security():
    # 1. Load all three sources
    users, settings, logs = load_files('users.json', 'security_settings.json', 'access_logs.json')

    restricted = settings['restricted_doors']

    print("--- Security Analysis Report ---")

    # 2. Loop through the logs (the "Events")
    for entry in logs:
        uid = entry['user_id']
        door = entry['door']
        time = entry['time']

        # 3. Lookup the name using the ID (Mapping)
        user_name = users.get(uid, "Unknown User")

        # 4. Check if the door is in the restricted list (Logic)
        if door in restricted:
            print(f"âŒ WARNING: {user_name} tried to enter the Restricted {door} door at {time}!")
        else:
            print(f"âœ… {time}: {user_name} accessed {door} (Authorized).")

analyze_security()
```

```python
import json

# Camera 1 (Main Entrance)
cam_1 = [
    {"person_id": "P1", "name": "Dr. Abbas"},
    {"person_id": "P2", "name": "Nurse Hussain"}
]

# Camera 2 (Medicine Storage - Restricted)
cam_2 = [
    {"person_id": "P1", "time": "10:00"},
    {"person_id": "P3", "time": "10:05"} # Notice: P3 was never at the Entrance!
]
with open('cam_1.json','w') as f:
  json.dump(cam_1,f,indent=2)
with open('cam_2.json','w') as f:
  json.dump(cam_2,f,indent=2)
```

```python
import json

# First, let's fix the loader to actually return the data
def load_files(filename):
    with open(filename, 'r') as f:
        return json.load(f)

def main_logic():
    # Load the data
    cam1_data = load_files('cam_1.json')
    cam2_data = load_files('cam_2.json')

    # INTERVIEW TIP: Convert cam1 to a dictionary for faster lookup
    # This maps 'P1' -> 'Dr. Abbas'
    authorized_people = {person['person_id']: person['name'] for person in cam1_data}

    print("--- Security Analysis Report ---")

    # We check everyone in Camera 2 (the restricted area)
    for entry in cam2_data:
        person_id = entry['person_id']
        detection_time = entry['time']

        # Check if this ID exists in our authorized dictionary
        if person_id in authorized_people:
            name = authorized_people[person_id]
            print(f"âœ… Authorized: {name} is in the restricted area.")
        else:
            # This is the "Security Gap"
            print(f"ðŸš¨ SECURITY BREACH: Unknown Person (ID: {person_id}) detected at {detection_time}!")

# Running the logic
main_logic()
```
