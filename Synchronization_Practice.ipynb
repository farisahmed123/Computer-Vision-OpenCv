```python
import json

def load_files(*filenames):
    return [json.load(open(f, 'r')) for f in filenames]

def analyze_security():
    # 1. Load all three sources
    users, settings, logs = load_files('users.json', 'security_settings.json', 'access_logs.json')

    restricted = settings['restricted_doors']

    print("--- Security Analysis Report ---")

    # 2. Loop through the logs (the "Events")
    for entry in logs:
        uid = entry['user_id']
        door = entry['door']
        time = entry['time']

        # 3. Lookup the name using the ID (Mapping)
        user_name = users.get(uid, "Unknown User")

        # 4. Check if the door is in the restricted list (Logic)
        if door in restricted:
            print(f"‚ùå WARNING: {user_name} tried to enter the Restricted {door} door at {time}!")
        else:
            print(f"‚úÖ {time}: {user_name} accessed {door} (Authorized).")

analyze_security()
```

```python
import json

# First, let's fix the loader to actually return the data
def load_files(filename):
    with open(filename, 'r') as f:
        return json.load(f)

def main_logic():
    # Load the data
    cam1_data = load_files('cam_1.json')
    cam2_data = load_files('cam_2.json')

    # INTERVIEW TIP: Convert cam1 to a dictionary for faster lookup
    # This maps 'P1' -> 'Dr. Abbas'
    authorized_people = {person['person_id']: person['name'] for person in cam1_data}

    print("--- Security Analysis Report ---")

    # We check everyone in Camera 2 (the restricted area)
    for entry in cam2_data:
        person_id = entry['person_id']
        detection_time = entry['time']

        # Check if this ID exists in our authorized dictionary
        if person_id in authorized_people:
            name = authorized_people[person_id]
            print(f"‚úÖ Authorized: {name} is in the restricted area.")
        else:
            # This is the "Security Gap"
            print(f"üö® SECURITY BREACH: Unknown Person (ID: {person_id}) detected at {detection_time}!")

# Running the logic
main_logic()
```

```python
import json

def load_file(filename):
    try:
        with open(filename, 'r') as f:
            return json.load(f)
    except FileNotFoundError:
        print(f"Error: {filename} not found!")
        return []

def main():
    eqp = load_file('equipment.json')

    low_battery_devices = []
    maintenance_count = 0

    print("--- Hospital Equipment Status Report ---")

    for device in eqp:
        # Use .get() to prevent crashes if a key is missing
        d_id = device.get('device_id', 'Unknown ID')
        battery = device.get('battery', 100)
        status = device.get('status', 'Normal')

        # 1. Logic for Maintenance
        if status == 'Maintenance Required':
            maintenance_count += 1
            print(f"üîß Alert: {d_id} requires maintenance.")

        # 2. Logic for Battery
        if battery < 20:
            low_battery_devices.append(d_id)

    print("-" * 40)
    print(f"Total Machines Needing Maintenance: {maintenance_count}")
    print(f"Critical Battery Levels (<20%): {', '.join(low_battery_devices) if low_battery_devices else 'None'}")

main()
```

```python
import json
config = {
  "thresholds": {
    "min_light_level": 30,
    "min_image_quality": 25,
    "max_motion_latency_ms": 2000,
    "high_motion_trigger": 7
  },
  "states": ["Normal", "Infrared Mode", "Manual Review Required", "Security Breach"]
}
monitoring_stream = [
  {
    "timestamp": 1706721000.5,
    "camera_id": "CAM_01",
    "person_id": "P_101",
    "signals": {
      "light_level": 45,
      "image_quality": 80,
      "motion_score": 2
    }
  },
  {
    "timestamp": 1706721005.2,
    "camera_id": "CAM_01",
    "person_id": "P_999",
    "signals": {
      "light_level": 15,
      "image_quality": 10,
      "motion_score": 8
    }
  },
  {
    "timestamp": 1706721012.8,
    "camera_id": "CAM_02",
    "person_id": "P_102",
    "signals": {
      "light_level": 50,
      "image_quality": 5,
      "motion_score": 0
    }
  }
]
identities = {
  "P_101": {"name": "Dr. Abbas", "access_level": "High"},
  "P_102": {"name": "Nurse Hussain", "access_level": "Standard"}
}
with open('config.json','w') as f:
  json.dump(config,f,indent=2)
with open('monitoring_stream.json','w') as f:
 json.dump(monitoring_stream,f,indent=2)
with open('identities.json','w') as f:
  json.dump(identities,f,indent=2)
```

```python
import json

def load_files(filename, default):
    try:
        with open(filename, 'r') as f:
            return json.load(f)
    except FileNotFoundError:
        return default

def main():
    config = load_files('config.json', {})
    monitoring_stream = load_files('monitoring_stream.json', [])
    identities = load_files('identities.json', {})

    thresholds = config.get('thresholds', {})
    last_timestamp = None  # We need this to calculate the "Sync" gap

    for i in monitoring_stream:
        # 1. SETUP DATA
        signals = i.get('signals', {})
        p_id = i.get('person_id')
        current_time = i.get('timestamp', 0)
        final_state = "Normal" # Default state for this specific log

        # 2. SYNC CHECK (Latency)
        if last_timestamp is not None:
            gap = current_time - last_timestamp
            if gap > 5: # If gap is more than 5 seconds
                print(f"‚ö†Ô∏è LATENCY WARNING: {gap:.1f}s delay between signals!")

        last_timestamp = current_time # Update for the next loop

        # 3. DECISION TREE (Logic)
        # Check Quality first (High Priority)
        if signals.get('image_quality', 0) < thresholds.get('min_image_quality', 0):
            final_state = "Manual Review Required"

        # Check Infrared fallback
        elif signals.get('light_level', 0) < thresholds.get('min_light_level', 0):
            final_state = "Infrared Mode"

        # Check Security Breach (Person not in identities AND high motion)
        if p_id not in identities and signals.get('motion_score', 0) > thresholds.get('high_motion_trigger', 7):
            final_state = "Security Breach"

        # 4. OUTPUT
        user_name = identities.get(p_id, {}).get('name', 'Unknown')
        print(f"[{current_time}] Person: {user_name} | State: {final_state}")

main()
```
